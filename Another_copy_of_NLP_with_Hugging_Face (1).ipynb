```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "The new policy aims to reduce carbon emissions significantly",
    candidate_labels=["environment", "health", "technology"],
)
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Matahari adalah bintang terdekat dengan Bumi dan merupakan pusat dari tata surya kita."
question = "Apa itu matahari?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Soekarno (lahir 6 Juni 1901 – meninggal 21 Juni 1970) adalah Presiden pertama Republik Indonesia. Ia menjabat dari tahun 1945 hingga 1967, dan dikenal sebagai salah satu proklamator kemerdekaan Indonesia."
question = "Siapa presiden pertama Republik Indonesia?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")
```

1. Zero-Shot Classification
Zero-shot classification adalah teknik di mana model dapat mengklasifikasikan data ke dalam label yang belum pernah dilihat sebelumnya selama pelatihan. Misalnya, jika model dilatih pada kategori seperti "olahraga" dan "politik," ia dapat mengklasifikasikan teks baru ke dalam kategori "seni" tanpa pelatihan khusus. Ini berguna dalam situasi di mana data label tidak tersedia atau sulit untuk dikumpulkan. Model menggunakan pemahaman bahasa untuk menafsirkan konteks dan memilih label yang paling sesuai.

2. Text Generation
Text generation melibatkan penggunaan model untuk menghasilkan teks yang koheren dan relevan secara kontekstual berdasarkan sebuah prompt atau kalimat awal. Ini bisa digunakan dalam berbagai aplikasi, seperti menulis kreatif, pembuatan konten, atau bahkan chatbot. Model seperti GPT-2 dan GPT-3 dari OpenAI sangat populer dalam tugas ini, mampu menghasilkan paragraf yang panjang dengan sedikit input.

3. Fill Mask
Fill-mask adalah teknik di mana model memprediksi kata atau frasa yang hilang dalam kalimat. Biasanya, kalimat diberikan dengan satu atau lebih kata yang diganti dengan token <mask>, dan model mencoba untuk mengisi tempat kosong tersebut. Ini berguna untuk memahami konteks dalam kalimat dan bisa digunakan dalam berbagai aplikasi, termasuk perbaikan kalimat atau pengayaan teks.

4. Named Entity Recognition (NER)
Named Entity Recognition adalah proses identifikasi dan klasifikasi entitas dalam teks. Entitas ini bisa berupa nama orang, organisasi, lokasi, tanggal, dan lainnya. NER penting dalam pemrosesan bahasa alami karena membantu dalam ekstraksi informasi dan pemahaman konten. Misalnya, dalam kalimat "Barack Obama lahir di Hawaii," NER akan mengidentifikasi "Barack Obama" sebagai orang dan "Hawaii" sebagai lokasi.

5. Question Answering
Question answering adalah kemampuan model untuk memberikan jawaban atas pertanyaan yang diajukan berdasarkan konteks atau teks tertentu. Model ini sering digunakan dalam sistem tanya jawab otomatis, di mana pengguna dapat mengajukan pertanyaan dan mendapatkan jawaban langsung dari dokumen yang relevan. Contohnya adalah model seperti BERT yang telah dilatih pada dataset tanya jawab, memungkinkan mereka untuk menangkap informasi penting dalam konteks.

6. Sentiment Analysis
Sentiment analysis adalah teknik yang digunakan untuk menentukan sikap atau perasaan penulis terhadap suatu topik dalam teks. Ini bisa berupa sentimen positif, negatif, atau netral. Misalnya, dalam analisis ulasan produk, model dapat membantu perusahaan memahami apakah pelanggan mereka puas atau tidak. Metode ini sering digunakan dalam analisis media sosial dan survei untuk menilai opini publik.

7. Summarization
Summarization adalah proses merangkum teks yang panjang menjadi versi yang lebih singkat sambil tetap mempertahankan informasi utama dan makna. Ada dua pendekatan untuk ini: ekstraktif (mengambil kalimat penting dari teks) dan abstraktif (menghasilkan ringkasan baru dengan kalimat yang berbeda). Ini sangat berguna dalam pembuatan laporan, artikel, atau pengelolaan informasi dalam organisasi.

8. Translation
Translation adalah tugas menerjemahkan teks dari satu bahasa ke bahasa lain. Model penerjemahan berbasis jaringan saraf, seperti model seq2seq, telah membawa kemajuan besar dalam kualitas terjemahan. Ini memungkinkan aplikasi seperti Google Translate untuk memberikan terjemahan yang lebih akurat dan alami. Penerjemahan otomatis sangat penting dalam komunikasi internasional dan pengembangan konten multibahasa.
